```python
excel_file = pd.ExcelFile('../data/RookiesData_cleaned.xlsx')
hojas = excel_file.sheet_names

bateadores_list = []
pitchers_list = []

for hoja in hojas:
    df_temp =pd.read_excel('../data/RookiesData_cleaned.xlsx', sheet_name=hoja)
    
    
    if 'Btt' in hoja:
        df_temp['Tipo'] = 'Bateador'
        bateadores_list.append(df_temp)
    elif 'Ptch' in hoja:
        df_temp['Tipo'] = 'Pitcher'
        pitchers_list.append(df_temp)
df_bateadores = pd.concat(bateadores_list, ignore_index=True)
df_pitchers = pd.concat(pitchers_list, ignore_index=True)

print("Bateadores Combinados:")
print(f"Filas: {df_bateadores.shape[0]}")
print(f"Años: {sorted(df_bateadores['DataY'].unique())}")
print("\nPitchers Combinados:")
print(f"Filas: {df_pitchers.shape[0]}")
print(f"Años: {sorted(df_pitchers['DataY'].unique())}")
```

```python
war_bateadores_equipo = df_bateadores.groupby('Tm')['WAR'].sum().sort_values(ascending=False)
print("WAR total por equipo (Bateadores):")
print(war_bateadores_equipo)
```

```python
war_pitchers_equipo = df_pitchers.groupby('Tm')['WAR'].sum().sort_values(ascending=False)
print("WAR acumulado por equipo (Pitchers):")
print(war_pitchers_equipo)
```

```python
war_combinado = pd.concat([
    df_bateadores[['Tm', 'WAR']],
    df_pitchers[['Tm', 'WAR']]
])

war_total_equipo = war_combinado.groupby('Tm')['WAR'].sum().sort_values(ascending=False)
print("WAR TOTAL acumulado por equipo (Bateadores + Pitchers):")
print(war_total_equipo)
```

```python
pitchers_calificados = df_pitchers[df_pitchers['IP'] >= 50]

print("Top 5 WAR:")
print(pitchers_calificados.nlargest(5, 'WAR')[['Name', 'WAR']])

print("\nTop 5 IP:")
print(pitchers_calificados.nlargest(5, 'IP')[['Name', 'IP']])

print("\nTop 5 ERA:")
print(pitchers_calificados.nsmallest(5, 'ERA')[['Name', 'ERA']])

print("\nTop 5 Ponches:")
print(pitchers_calificados.nlargest(5, 'SO')[['Name', 'SO']])

print("\nTop 5 WHIP:")
print(pitchers_calificados.nsmallest(5, 'WHIP')[['Name', 'WHIP']])

print("\nTop 5 Salvados:")
print(pitchers_calificados.nlargest(5, 'SV')[['Name', 'SV']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
plt.Figure(figsize=(14, 7))

ax = war_total_equipo.plot(kind='bar', color='gray')
ax.bar_label(ax.containers[0], fmt='%.1f', fontsize=9, padding=3)
plt.title('WAR Total de Novatos por Equipo (Bateadores + Pitchers) 2020-2025', fontsize=14)
plt.xlabel('Equipo')
plt.ylabel('WAR Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
comparativo = pd.DataFrame({'Bateadores': war_bateadores_equipo, 'Pitchers': war_pitchers_equipo}).fillna(0)
comparativo = comparativo.sort_values(by='Bateadores', ascending=False)

ax = comparativo.plot(kind='bar', stacked=False, figsize=(18, 7), color=['skyblue', 'salmon'], width=0.7)

# Etiquetas de Bateadores (ocultando valores casi en cero)
labels_bateadores = [f'{v:.1f}' if abs(v) > 0.05 else '' for v in ax.containers[0].datavalues]
ax.bar_label(ax.containers[0], labels=labels_bateadores, fontsize=9, padding=3)

# Etiquetas de Pitchers (ocultando valores casi en cero)
labels_pitchers = [f'{v:.1f}' if abs(v) > 0.05 else '' for v in ax.containers[1].datavalues]
ax.bar_label(ax.containers[1], labels=labels_pitchers, fontsize=9, padding=3)

plt.title('Comparativo de WAR de Novatos por Equipo (Bateadores vs Pitchers) 2020-2025', fontsize=14)
plt.ylabel('WAR Total')
plt.xlabel('Equipo')
plt.xticks(rotation=45)
plt.legend(title='Tipo')
plt.tight_layout()
plt.show()
```
